```python
# YOUR CODE HERE
plt.hist(dataset['co2'],edgecolor='black',bins=range(0, 600, 25))
plt.xlabel('co2 amount')
plt.xticks(range(0, 650, 50))
plt.ylabel('co2 frequency')
plt.show()
```

The above diagram shows that most co2 emissions are between 200 and 550. The distribution also shows a sudden stop around 575.

```python
# YOUR CODE HERE
plt.hist(dataset['VClass'],edgecolor='black')
plt.title('Car Class distribution')
plt.ylabel('number of cars')
plt.show()
```

The histogram above show Class distributions among cars. There Midsize and Compact cars are the most prominent(around 1190 and 1170 respectively.) Mini compact cars being the least at roughly 300. The most prominent are about 4 times as the least popular (Minicompact) cars.

```python
# YOUR CODE HERE
plt.hist(dataset['fuelType'],edgecolor='black')
plt.title('Fuel Type distribution')
plt.show()
```

Fuel Type distribution per this histogram tells us that Gasoline is virtually the dominant fuel type by a large margin. The minority is comprised of Diesel, Midgrade Gasoline (despite being gasoline too), and natural gas. Premium Gasoline take the lead and is about 3/2 of the Regulara Gasoline (second most dominant.)
